```python
year= 2016
variable = 'Individuals using the Internet (% of population)' 
```

```python
data_year = data[data['Year'] == year]
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data_year[variable], bins=20, edgecolor='m', alpha=0.7)
plt.title(f'Distribution of {variable} in {year}')
plt.xlabel(variable)
plt.ylabel('Frequency')
plt.show()
```

```python
plt.title(f'Distribution of {variable} in {year}')
plt.xlabel(variable)
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.7)
```

```python
region_counts = data['Region'].value_counts()

# Create a bar chart
plt.figure(figsize=(10, 6))
region_counts.plot(kind='bar', color='purple')
plt.xlabel('Region')
plt.ylabel('Number of Countries')
plt.title('Distribution of Countries by Region')
plt.xticks(rotation=45) 
plt.show()
```

```python
plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 2)
region_percentage = (region_counts / region_counts.sum()) * 100
colors = ['maroon', 'pink', 'red', 'plum']
region_percentage.plot(kind='pie', autopct='%1.1f%%', colors=colors)
plt.axis('equal')  
plt.title('Distribution of Countries by Region (Percentage)')
plt.tight_layout()
plt.show()
```
